```python
act = act.stack().reset_index(level=1,drop=True).to_frame()
```

```python
act
```

```python
act.columns=['Actor']
```

```python
act
```

merge the dataframe with actors names and other columns

```python
act=act.merge(df[['title','revenue_musd','vote_average','popularity']],how='left',left_index=True, right_index=True)
```

```python
act
```

```python
act.Actor.unique()
```

```python
act.Actor.value_counts().head(20)
```

```python
plt.figure(figsize=(12,8))
act.Actor.value_counts().head(20).plot(kind='bar',fontsize=15)
#top 20 actor worked in number of films
plt.title("Top 20 Most Frequent Actors",fontsize=20)
plt.ylabel("Number of Movies",fontsize=15)
plt.show()
```

created aggregate by columns

```python
agg=act.groupby('Actor').agg(Total_Revenue=('revenue_musd','sum'),
                             Mean_Revenue=('revenue_musd','mean'),
                             Mean_Rating=('vote_average','mean'),
                             Mean_pop=('popularity','mean'),
                          
                             Total_Movies=('Actor','count')
                             )
```

```python
agg.nlargest(10,'Total_Movies')
```

```python
agg.nlargest(10,'Total_Revenue')
```

```python
plt.figure(figsize=(12,8))
agg.Total_Revenue.nlargest(20).plot(kind='bar',fontsize=15)
plt.title ("Total Revnue",fontsize=20)
plt.ylabel("Revenue in Million USD",fontsize=15)

plt.show()
```
